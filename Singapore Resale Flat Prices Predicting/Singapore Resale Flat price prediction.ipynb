```python
import pandas as pd
dv = pd.read_csv('combined.csv')
dv.shape
```

```python
dv['month'] = dv['month'].str.split('-').str[0]
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Repeat the process for other columns if needed
dv['flat_type_encoded'] = label_encoder.fit_transform(dv['flat_type'])
dv['flat_model_encoded'] = label_encoder.fit_transform(dv['flat_model'])
dv['town encoder'] = label_encoder.fit_transform(dv['town'])
dv['street_name_encoder'] = label_encoder.fit_transform(dv['street_name'])
```

```python
dv['block_encoder'] = label_encoder.fit_transform(dv['block'])
```

```python
# Function to extract and encode the storey_range
def encode_storey_range(range_str):
    start, end = map(int, range_str.split(' TO '))
    midpoint = (start + end) / 2
    return midpoint

# Apply the function to create a new column 'Encoded'
dv['Encoded storey range'] = dv['storey_range'].apply(encode_storey_range)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = dv.drop(['resale_price'], axis = 1)
y = dv['resale_price']
X_train, X_test, y_train, y_test =  train_test_split(X,y,test_size=0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
y_pred = lr.predict(X_test)
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R-squared (R2) Score: {r2}")
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()

# Fit the model to the training data
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R-squared (R2) Score: {r2}")
```

```python
regressor.score(X_test, y_test)
```
